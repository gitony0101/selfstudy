```python

from numpy.random import default_rng
rng = default_rng(5)
import numpy as np
n = 10
x = rng.uniform(low=0.0,high=10.0,size=n)
x_mean = np.mean(x)
x_var = np.var(x)

print("the random numebrs are",x)
print("the mean is",x_mean)
print("the variance is",x_var)
```

```python
from numpy.random import default_rng
rng = default_rng(5)

mu, sigma = 5,10
n = 5
x = rng.normal(mu,sigma,n)
x_mean = np.mean(x)
x_var = np.var(x)
print("the random numebrs are",x)
print("the mean is",x_mean)
print("the variance is",x_var)
```
