```python
x, y = 3,4
if (x > y):
    print(x,'>', y)
elif(x ==y):
    print(x,'equals', y)
else:
    print('Either', x, '<', y, 'or x and y are not orderable')
```

```python
fruits = ['apple', 'banana', 'cherry', 'durian', 'eggplant', 'fig']
'durian' in fruits
```

```python
# for
for fruit in fruits:
    print(fruit)
```

```python
# while
i = 0
while (i<5):
    print(i)
    i += 1
```

continue and break

```python
for i in range(2):
    for j in range(2):
        for z in range(2):
            if i==j==z:
                continue
            print(i,j,z)
```

```python
i = 0 
while True:
    print(i)
    if i >5:
        break
    i += 1
```
