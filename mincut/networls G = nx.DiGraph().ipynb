```python
import networkx as nx # NetworkX v1.10


#   >c->g
G = nx.DiGraph()
G.add_edge("a", "1", capacity=2)
G.add_edge("a", "2", capacity=2)
G.add_edge("a", "3", capacity=2)
G.add_edge("a", "4", capacity=2)

G.add_edge("1", "2", capacity=2)
G.add_edge("1", "6", capacity=2)
G.add_edge("1", "5", capacity=2)

G.add_edge("2", "6", capacity=2)
G.add_edge("2", "3", capacity=2)

G.add_edge("3", "4", capacity=2)
G.add_edge("3", "7", capacity=2)

G.add_edge("4", "7", capacity=2)
G.add_edge("4", "8", capacity=2)

G.add_edge("5", "6", capacity=2)
G.add_edge("5", "b", capacity=2)

G.add_edge("6", "b", capacity=2)
G.add_edge("6", "7", capacity=2)

G.add_edge("7", "b", capacity=2)
G.add_edge("7", "8", capacity=2)


cut_nodes = nx.algorithms.connectivity.minimum_node_cut(G, "a", "b")
print("Cut nodes: ",cut_nodes)
```
