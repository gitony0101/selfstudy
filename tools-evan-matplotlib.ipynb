```python
x = np.linspace(-1.2,1.2,300)
plt.plot(x,x**2,'r:',x,x**3,'b:',x,np.sin(x),'g:')
plt.show()
```

```python
# line1, line2, line3 

x = np.linspace(-1.2,1.2,300)
line1, line2, line3  = plt.plot(x,x**2,'r:',x,x**3,'b:',x,np.sin(x),'g:')
line1.set_linewidth(3.0)
line1.set_dash_capstyle("projecting")
line3.set_alpha(0.9)#透明度

plt.show()
```

```python
# save

x = np.linspace(-1.2,1.2,100)
line1, line2, line3  = plt.plot(x,x**2,'r:',x,x**3,'b:',x,np.sin(x),'g:')
line1.set_linewidth(2.0)
line1.set_dash_capstyle("butt")
line3.set_alpha(0.9)#透明度
plt.grid(True)
plt.savefig("my square func grid.png")
plt.show()
```

### subpolts

```python
x = np.linspace(-1,1,30)

plt.subplot(2,2,1) # top left
plt.plot(x,x)

plt.subplot(2,2,2) # top right
plt.plot(x,x**2)

plt.subplot(223) # bottom left 妈蛋简写也行！！！！223！！！！
plt.plot(x,np.sin(x))

plt.subplot(2,2,4) # bottom right
plt.plot(x,x**3)

plt.show()
```

```python
x = np.linspace(-1,1,30)

plt.subplot(2,2,1) # top left
plt.plot(x,x)

plt.subplot(2,2,2) # top right
plt.plot(x,x**2)

plt.subplot(212) # bottom left 妈蛋简写也行！！！！223！！！！
plt.plot(x,np.cos(x))


plt.show()
```

```python
# more complex
plt.subplot2grid((3,3), (0, 0), rowspan=2, colspan=2)
plt.plot(x, x**2)
plt.subplot2grid((3,3), (0, 2))
plt.plot(x, x**3)
plt.subplot2grid((3,3), (1, 2), rowspan=2)
plt.plot(x, x**4)
plt.subplot2grid((3,3), (2, 0), colspan=2)
plt.plot(x, x**5)
plt.show()
```

If you need even more flexibility in subplot positioning, check out the [GridSpec documentation](http://matplotlib.org/users/gridspec.html)
