```python
import numpy as np

x = np.linspace(-2,2,500)

y = x ** 2

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("square function")
plt.xlabel("x")
plt.ylabel("y=x**2")
plt.grid(True)
plt.show()
```
